```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton

app = QApplication(sys.argv)

window = QWidget()
window.setWindowTitle("Hello PyQt")
window.setGeometry(100, 100, 300, 200)

button = QPushButton("Click me", window)
button.move(100, 80)

window.show()

sys.exit(app.exec_())
```
